### Boolean Masken

Vektorisierung mit Booleschen Operatoren

Wird für Filterung verwendet

!!WICHTIG!!

```python
f = np.arange(20)
```

```python
f
```

```python
f % 2 == 0  # Boolean Array
```

```python
# Wenn jetzt dieses Ergebnisarray auf das Original vektorisiert wird, kommen nur die Zahlen heraus, die mit True kombiniert werden
x = f % 2 == 0
f[x]
```

```python
f > 10
```

```python
f[f % 2 == 0]  # Kurzform (ohne Variable)
```

Beispiel

```python
y = np.random.randint(10, size=(20))
```

```python
y
```

```python
y % 2 == 0
```

```python
y[y % 2 == 0]
```

```python
np.arange(len(y))[y % 2 == 0]  # Stellen finden von den geraden Zahlen
```

Beispiel 2

```python
# Aufgabe: In y alle überdurchschnittlichen Werte finden
avg = y.mean()  # Durchschnitt finden
y > avg  # Boolean Maske
z = y > avg
y[z]  # Filterung
```

```python
y[y > y.mean()]  # Kurzform
```

Filterung mit Matrix

```python
c
```

```python
# Aufgabe: Alle Zeilen finden, die überdurchschnittlich sind
gesamtAvg = c.mean()
zeilenAvg = c.mean(axis=1)
zeilenAvg[zeilenAvg > gesamtAvg]
```

```python
zeilenAvg
```

hstack, vstack

Numpy Array an ein Datenset rechts/unten anhängen

Beispiel: Summenzeile/Summenspalte

```python
s = np.random.randint(100, size=(5, 5))
```

```python
s
```

```python
# Aufgabe: Rechts + Unten eine Summe anhängen
s.sum(axis=0)
```

```python
v = np.vstack([s, s.sum(axis=0)])  # WICHTIG: Bei vstack/hstack müssen die Parameter in einer Liste sein
```

```python
v
```

```python
v.sum(axis=1)  # hstack kann nur 2D-Arrays anhängen -> reshape
```

```python
v.sum(axis=1).reshape(-1, 1)
```

```python
np.hstack([v, v.sum(axis=1).reshape(-1, 1)])
```
