### Spalten bearbeiten

```python
data.drop(columns=["#"], inplace=True)  # Hier können auch Zeilen gelöscht werden
```

```python
previousNames = data.columns

names = {
    "Population(2020)": "Pop",
    "Density(P/Km²)": "Density",
    "Land Area(Km²)": "LandArea",
    "Migrants(net)": "MigrantsNet",
    "Fert.Rate": "FertRate",
    "Med.Age": "MedAge",
    "UrbanPop %": "UrbanPop"
}

data.rename(columns=names, inplace=True)  # Hier können auch Zeilen umbenannt werden
```

```python
data
```

```python
previousNames
```

```python
data.columns = previousNames  # Vorherige Namen wiederherstellen
```

```python
data
```

#### Sortieren
- sort_index
- sort_values(Spaltenname(n))

#### Daten filtern
- Normaler Index (data[...]): Spaltenfilterung
- loc: Zeilenfilterung
- iloc: Indexloc, Zeilenfilterung nach Zahlen

```python
data["Pop"]
```

```python
data["Pop"] > 10_000_000
```

```python
data[data["Pop"] > 10_000_000]
```

```python
data[(data["Pop"] > 10_000_000) & (data["Density"] > 1000)]  # Bei mehreren Filterkriterien muss & und | verwenden (nicht and und or), alle Terme müssen eingeklammert sein
```

```python
data.loc["China":"Pakistan"]
```

```python
data.loc["China"]
```

```python
data.loc["China":"Pakistan"]["Pop"]  # Einzelne Spalte aus einer Zeilenmenge
```

```python
data.loc["China":"Pakistan", "Pop":"NetChange"]  # Nur bestimmte Spalten auswählen
```

```python
data.loc[:, "Pop": "NetChange"]  # Alle Zeilen mit bestimmten Spalten
```

```python
data.loc[:, ("Pop", "WorldShare")]  # Bestimmte Spalten entnehmen mit ( ) -> Liste
```

```python
data.index[data.index.str.contains("A") | data.index.str.contains("B")]  # Spalte nach allen Texten suchen die ein A oder B enthalten
```

```python
data.loc[data.index[data.index.str.contains("A") | data.index.str.contains("B")]]  # Finde alle Datensätze die A oder B enthalten (beim Country)
```

```python
data["MedAge"].value_counts()  # Anzahlen von identischen Werten pro Spalte suchen
```

```python
data.value_counts()  # Alle Zeilen auf Eindeutigkeit prüfen
```

#### Gruppierung

Datensätze anhand eines Kriteriums in Gruppen aufteilen

Danach können diese Gruppen separat analysiert werden, oder auf jede Gruppe eine Auswertung ausführen

```python
data.groupby("MedAge")
```

```python
data.groupby("MedAge").get_group("19")["Pop"].mean()  # Einzelne Gruppe auswerten
```

```python
data.groupby("MedAge")["Pop"].mean()  # Alle Gruppen auswerten
```

```python
for x in data.groupby("MedAge")["Pop"].mean():  # Absolute Werte ausgeben
    print(int(x))
```
