### NaN befüllen
- manuell
- ffill(), bfill()

```python
data.ffill()  # Forwardfill, nimmt den letzten Wert (von oben) und kopiert diesen hinunter; bfill nimmt den letzten Wert von unten
```

```python
data.bfill(inplace=True)  # Änderungen übernehmen
```

#### Manuelles befüllen

Spalte auswählen, Null-Werte innerhalb dieser Spalte finden, Null-Werte mit = befüllen, Spalten zurückschreiben ins Datenset

```python
b = data["MigrantsNet"]
```

```python
b.isna()  # Filterung der alleinigen Spalte nach Null-Werten
```

```python
b[b.isna()]  # Filterung auf die Spalte selbst anwenden
```

```python
b[b.isna()] = 0  # Neuen Wert hineinschreiben
```

```python
b
```

```python
data["MigrantsNet"] = b
```

```python
data
```

```python
# Eigene Funktion schreiben
def fixNaN(columnName: str):
    col = data[columnName]
    b = col[col.isnull()]
    col[b] = 0
    data[columnName] = col
```

```python
fixNaN("LandArea")  # Funktion kann jetzt mit Spaltenname einfach verwendet werden
```

### Prozentzeichen

Hier wird die string-Replace Funktion verwendet

```python
data["YearlyChange"].replace(" %", "", regex=True)  # Hier muss regex=True verwendet werden (weil keine Ahnung)
```

```python
data["YearlyChange"] = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

```python
def fixPct(columnName: str):
    data[columnName] = data[columnName].replace(" %", "", regex=True)
```

```python
data.columns
```

```python
for col in data.columns:  # Aus jeder Spalte % entfernen
    fixPct(col)
```

```python
data
```

### N.A.
Selbes Schema wie Prozentzeichen

```python
def fixNA(columnName: str, newValue):
    data[columnName] = data[columnName].replace("N.A.", newValue)
```

```python
for col in data.columns:  # Aus jeder Spalte N.A. entfernen
    fixNA(col, "0")
```

```python
data
```

### Datentypen

Bei der Behebung von Fehlern werden die Typen nicht automatisch angepasst

Gute Typen verkleinern das Datenset und erhöhen die Performance

```python
data.info()
```

```python
data.convert_dtypes().info()  # MigrantsNet float -> Int64
```

convert_dtypes ist nicht sonderlich akkurat -> manuell

astype(Typ): Forciert eine Konvertierung; im Vorhinein muss klar sein, welcher Typ optimal ist

```python
data
```

```python
data["Pop"] = data["Pop"].astype(np.int32)
```

```python
def fixType(columnName: str, newType):
    data[columnName] = data[columnName].astype(newType)
```

```python
fixType("YearlyChange", np.float32)
```

```python
fixType("NetChange", np.int32)
```

```python
fixType("Density", np.int16)
```
