# Datenbank

```python
import pyodbc
import numpy as np
import pandas as pd
```

```python
connStr = "Driver={ODBC Driver 17 for SQL Server};Server=WIN11-LK3;Database=Northwind;Trusted_Connection=yes;"
with pyodbc.connect(connStr) as conn:
    cursor = conn.cursor()
```

```python
cursor = conn.cursor()

cursor.execute("SELECT * FROM Customers")

columns = [column[0] for column in cursor.description]
rows = cursor.fetchall()
```

```python
cursor.description
```

```python
rows
```

```python
pd.DataFrame.from_records(rows, columns=columns)
```

```python
pd.read_sql("SELECT * FROM Customers", con=conn)
```

# Enumerate

```python
data = pd.read_csv("Data/PopulationDataFertig.csv")
```

```python
i = 0
for country in data["Country"]:
    print(f"{i}: {country}")
    i += 1
```

```python
for i, country in enumerate(data["Country"]):  # enumerate: Fügt zu einer for-Schleife einen Index hinzu
    print(f"{i}: {country}")
```
