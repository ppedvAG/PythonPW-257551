## Shortcuts
- Shift/Strg + Enter: Zelle ausführen
- Enter/Escape: Zelle betreten & verlassen
- A, B: Neue Zelle darüber/darunter einfügen
- D + D: Zelle löschen
- X, C, V: Cut, Copy, Paste
- Strg + Shift + Pfeiltaste: Zellen verschieben
- M, Y: Zwischen Markdown und Code umschalten
