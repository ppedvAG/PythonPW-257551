## Globale vs. Objektorientierte Graphen

Bei Globalen Graphen werden die entsprechenden Matplotlib Befehle verwendet, und der Plot wird im Anschluss angezeigt

Bei Objektorientierten Graphen werden die Plots in Variablen geschrieben, und über die Variablen wird dann die Konfiguration durchgeführt

```python
# Global
plt.figure(figsize=(15, 3))
plt.plot(data["Close"])
plt.xlabel("Jahr")
plt.ylabel("Preis USD")
plt.show()  # show(): Schließt die Konfiguration ab, und zeigt den Plot an

plt.plot(data["Open"])
plt.xlabel("Jahr")
plt.ylabel("Preis USD")
plt.show()
```

```python
# Objektorientiert
fig, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(2, 2)  # Bei den Variablen muss hier ein Tupel von Tupeln verwendet werden
a1_1.bar(data.index[:10], data.iloc[:10, -1])

a1_2.plot(data["Close"], color="red")

a2_1.pie(np.arange(5))

a2_2.plot(data["Volume"])

None
```
